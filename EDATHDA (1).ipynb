```python
import pandas as pd
import numpy as np
```

```python
# read data
url = "https://raw.githubusercontent.com/salemprakash/EDA/main/Data/1000%20Sales%20Records.csv"
df  = pd.read_csv('https://raw.githubusercontent.com/salemprakash/EDA/main/Data/1000%20Sales%20Records.csv')
```

```python
# Check the dimensions of the dataset
print("Dimensions:", df.shape)

# Get a summary of the dataset
print(df.info())

# Summary statistics
print(df.describe(include='all'))
```

```python
# Check for missing values
print(df.isnull().sum())

# Check for duplicate records
print("Duplicate records:", df.duplicated().sum())
```

```python
# Fill missing values or drop them
df = df.dropna()

# Remove duplicates
df = df.drop_duplicates()
```

```python
# Perform calculations on the 'Units Sold' column
total_sales = df['Units Sold'].sum()
mean_sales = df['Units Sold'].mean()
average_sales = df['Units Sold'].mean()  # The mean is typically referred to as the average

# Display the results
print("Total Units Sold:", total_sales)
print("Mean Units Sold:", mean_sales)
print("Average Units Sold:", average_sales)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histograms for numerical features
df.hist(bins=15, figsize=(15, 10))
plt.show()

# Count plots for categorical features
categorical_cols = ['Item Type', 'Region']
for col in categorical_cols:
    sns.countplot(df[col])
    plt.title(col)
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Check the data types of each column
print(df.dtypes)

# Select only numeric columns for correlation analysis
numeric_data = df.select_dtypes(include=['float64', 'int64'])

# Check if there are any numeric columns
if not numeric_data.empty:
    # Calculate the correlation matrix
    correlation_matrix = numeric_data.corr()
    # Plot the heatmap
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
    plt.title("Correlation Heatmap")
    plt.show()
else:
    print("No numeric columns found for correlation analysis.")
```

```python
# Box plots for numerical vs categorical
sns.boxplot(x='Item Type', y='Units Sold', data=df)
plt.title("Units Sold by Item Type")
plt.show()
```

```python
# Pairplot to visualize relationships
sns.pairplot(df)
plt.show()

# Using FacetGrid for categorical variables
g = sns.FacetGrid(df, col="Item Type", hue="Region")
g.map(sns.scatterplot, "Units Sold", "Total Revenue")
plt.show()
```
